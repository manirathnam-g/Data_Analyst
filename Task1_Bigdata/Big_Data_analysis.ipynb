```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg, max, min

from google.colab import files
uploaded = files.upload()

import os
os.listdir()


# 1. Create Spark session
spark = SparkSession.builder \
    .appName("BigDataAnalysis") \
    .getOrCreate()

# 2. Load data
df = spark.read.csv("sale.csv", header=True, inferSchema=True)

# 3. Show schema
df.printSchema()

# 4. Basic stats
df.describe().show()

# 5. Example analysis: top categories or value counts
df.groupBy("category_column").count().orderBy("count", ascending=False).show()

# 6. Example: average by group
df.groupBy("region").agg(avg("sales").alias("avg_sales")).show()

# 7. Cleanup
spark.stop()
```

```python
!pip install dask[complete]

import dask.dataframe as dd

# Load the CSV file
df = dd.read_csv('sale.csv')

# Show schema
df.dtypes

# Basic statistics
df.describe().compute()

# Value counts (top categories)
df['category_column'].value_counts().compute()

# Average sales by region
df.groupby('region')['sales'].mean().compute()

```

```python
import pandas as pd

# Load original file and repeat data
original = pd.read_csv("sale.csv")
large_data = pd.concat([original] * 100000)  # Creates a dataset with 1M rows

# Save and reload with Dask
large_data.to_csv("Large_Sales.csv", index=False)
df_large = dd.read_csv("Large_Sales.csv")

# Do groupBy or describe on the large dataset
df_large.groupby('region')['sales'].mean().compute()
```
